```python
 def plot_best(yhat, y):
      
      plt.plot( y.index, y)
      plt.plot( y.index, yhat)
      plt.legend(["Ground Truth",  "Prediction"])
      plt.title("The Plot of Predictions and Ground Truth versus Date")
      plt.xlabel("Date")
      plt.ylabel("Energy Demand (kWh)")
      plt.show()
      return None
```

```python
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(penalty = "l1", alpha = 0.0005, shuffle=False, learning_rate="constant", eta0 =0.018)
model.fit(X_train, Y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
print(mse(Y_train,(y_train_pred)))
print(mae(Y_train,(y_train_pred)))
print(r2(Y_train,(y_train_pred).reshape(1,-1)))
print("\n")
print(mse(Y_test,(y_test_pred)))
print(mae(Y_test,(y_test_pred)))
print(r2(Y_test,(y_test_pred.reshape(1,-1))))
```

```python
plot_best(y_test_pred,Y_test)
```
